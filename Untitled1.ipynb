```python
import numpy as np
import matplotlib.pyplot as plt

def g_n(x, N):
    result = np.zeros_like(x)
    for n in range(1, N + 1):
        result += (2 * (-1)**n * (1 - np.pi**2 * n**2 + 6) / n**3) * np.sin(n * x)
    return result

# Valores de x
x = np.linspace(0, 2 * np.pi, 1000)

# Nmero de trminos
N_values = [2, 5, 20, 1000]

# Graficar
plt.figure(figsize=(10, 6))
for N in N_values:
    plt.plot(x, g_n(x, N), label=f'N={N}')

plt.plot(x, x**3, label='$x^3$', linestyle='dashed', color='black')

plt.xlabel('x')
plt.ylabel('g_N(x)')
plt.legend()
plt.title('Convergencia de g_N(x) a $x^3$')
plt.grid(True)
plt.show()
```
